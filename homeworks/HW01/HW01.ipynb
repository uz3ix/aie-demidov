```python
arr = np.array([[1,2],[3,4]])
brr = np.array([[5,6],[7,8]])

np.vstack((arr,brr)), np.hstack((arr,brr)) # объединяет массивы по вертикальным осям, объединяет массивы по горизонтальным осям соответственно
```

```python
np.column_stack((arr, brr)), np.row_stack((arr,brr)) #первый вариант по горизонтальным осям, второй вариант устарел
```

```python
data = np.arange(1, 11)  

print("Сумма элементов:", data.sum())
print("Среднее значение:", data.mean())
print("Максимум:", data.max())
print("Минимум:", data.min())
print("Стандартное отклонение:", data.std())


matrix = np.random.randint(1, 10, (3, 4))
print(matrix)

print("Сумма по всем элементам:", matrix.sum())
print("Среднее по всем элементам:", matrix.mean())
print("Сумма по строкам:", matrix.sum(axis=1))
print("Сумма по столбцам:", matrix.sum(axis=0))
```
